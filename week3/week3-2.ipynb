```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/gdrive/MyDrive/인공지능/3주차/pima-indians-diabetes.csv',
names = ["pregnant", "plasma", "pressure", "thickness", "insulin", "BMI", "pedigree", "age", "class"])
```

```python
mean_age = df['age'].mean()
print('Average age: {:.2f}'.format(mean_age))

low_age = df[df['age'] <= mean_age]['thickness']
high_age = df[df['age'] > mean_age]['thickness']

import matplotlib.pyplot as plt

low_age.plot.hist(bins=50, alpha=0.5)
high_age.plot.hist(bins=50, alpha=0.7)
plt.legend(['low_age', 'high_age'], loc="upper right")
plt.xlabel('thickness')
plt.show();
```
